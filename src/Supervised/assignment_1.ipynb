## Naive approach

### Separating data into train and test

```python
predictors, target = titanic_dataset.iloc[:, 1:], titanic_dataset.iloc[:, :1]
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, random_state=123)
target_train.head()
```

### Training tree clasifier model

```python
clf = tree.DecisionTreeClassifier()
clf.fit(predictors_train, target_train)
```

### What happened
The error we are seeing is where sklearn is trying to convert a string value into a float.
Models tend to prefer a single type of input data. In our case, we have strings, floats and int datatypes.
What we should do is instead turn everything that we can into floats

## Less naive approach

### Preparing data

```python
titanic_dataset_preprocessed = titanic_dataset.copy(deep=True)
```

```python
titanic_dataset_preprocessed["Sex"] = (titanic_dataset_preprocessed["Sex"]=="male").astype("float64")
```

```python
titanic_dataset_preprocessed.describe()
```

```python
titanic_dataset_preprocessed.head()
```

```python
print("Number of unique ticket values: ", titanic_dataset_preprocessed["Ticket"].nunique())
print("Number of unique cabin values: ", titanic_dataset_preprocessed["Cabin"].nunique())
print("Number of unique embarked values: ",titanic_dataset_preprocessed["Embarked"].nunique())
```

```python
titanic_dataset_preprocessed = titanic_dataset_preprocessed.drop(["Name", "Ticket", "Cabin"], axis=1)
```

```python
titanic_dataset_preprocessed
```

```python
titanic_dataset_preprocessed = pd.get_dummies(titanic_dataset_preprocessed).astype("Float64").dropna()
```

### Separating data into test and train

```python
predictors, target = titanic_dataset_preprocessed.iloc[:, 1:], titanic_dataset_preprocessed.iloc[:, :1]
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, random_state=123)
predictors_train.head()
```

```python
target_test.head()
```

### Decision Tree classifier

```python
model = tree.DecisionTreeClassifier(max_leaf_nodes=5, max_depth=5)
model.fit(predictors_train, target_train)
```

### Stochastic gradient descent

```python
model = SGDClassifier("log_loss",verbose=True, early_stopping=False, shuffle=True, random_state=5, learning_rate="adaptive")
model.fit(predictors_train, target_train)
```

### Accuracy

```python
target_test.iloc[:, 0]
# predictors_test.iloc[1, :]
```

```python
# predictions = np.zeros(predictions.shape, np.float64)
predictions = model.predict(predictors_test)
```

```python
predictions
```

```python
n = 0
correct = 0
for prediction in predictions:
    correct += int(prediction==target_test.iloc[n, 0])
    n += 1

print("Accuracy rate: ", (correct)/n)
print("Total number of guesses: ", n)
```

```python
true_positive = 0
true_negative = 0
false_positive = 0
false_negative = 0
predictions = clf.predict(predictors_test)

for i, prediction in enumerate(predictions):
    if prediction == target_test.iloc[i, 0]:
        if prediction:
            true_positive += 1
        else:
            true_negative += 1
    else:
        if prediction:
            false_positive += 1
        else:
            false_negative += 1

print("Accuracy rate: ", (correct)/n)
print("Total number of guesses: ", n)
print("Confusion matrix:")
print(f"True positive: {true_positive} \t False negative: {false_negative}")
print(f"False positive: {false_positive} \t True negative: {true_negative}")
```
